```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
df = pd.read_csv('data.csv')
df = df[df['year']<2016]
df
```

```python
df.groupby('year')['suicides_no'].mean().plot(figsize=(12,8))
```

```python
df.groupby('year')['suicides/100k pop'].mean().plot(figsize=(12,8))
```

```python
plt_df = df.groupby(by=['year', 'age'])['suicides_no'].sum().unstack(-1)
plt_df = plt_df[['5-14 years', '15-24 years', '25-34 years', '35-54 years', '55-74 years', '75+ years']]
plt_df.plot.area(figsize=(12,8))
```

```python
plt_df = df.groupby(by=['year', 'age'])['suicides/100k pop'].mean().unstack(-1)
plt_df = plt_df[['5-14 years', '15-24 years', '25-34 years', '35-54 years', '55-74 years', '75+ years']]
plt_df.plot.area(figsize=(12,8))
```

```python
plt_df = df.groupby(by=['year', 'generation'])['suicides/100k pop'].mean().unstack(-1)
plt_df = plt_df[['G.I. Generation', 'Silent', 'Boomers', 'Generation X', 'Millenials', 'Generation Z']]
plt_df.plot(figsize=(12,8))
```

```python
plt_df = df.groupby(by=['year', 'sex'])['suicides/100k pop'].mean().unstack(-1)
plt_df.plot(figsize=(12,8))
```
